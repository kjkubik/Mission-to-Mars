```python
# Import Splinter, BeautifulSoup, and Pandas
from splinter import Browser
from bs4 import BeautifulSoup as soup
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
import re
```

```python
# Set the executable path and initialize Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# 1. Use browser to visit the URL, Astrogeology USGS GOV:
url = 'https://astrogeology.usgs.gov/maps/mars-viking-hemisphere-point-perspectives/'
browser.visit(url)
url
# NOTE: THE ORIGINAL SITE WAS DOWN SO I USED THE ONE ABOVE
```

```python
# 3. Write code to retrieve the image urls and titles for each hemisphere (THE FOUR THAT ARE 'ENHANCED')
```

```python
# Parse the resulting html with soup
html = browser.html
products_soup = soup(html, 'html.parser')
products_soup
```

```python
# Finding titles 
products_block = products_soup.find('div', class_='widget block products')
product_title = products_block.find_all('h3')
#print(product_title)

hemisphere_image_urls = []

for title in product_title:
    titles = title.text
    hemisphere_image_urls.append(titles)
print(hemisphere_image_urls)


#for products in products_titles:
#    product_string = products.string
#    print(product_string)
#    if products.string == re.search("%Enhanced").string:
#       products_list.append({products})
#print(products_list)
         #print (products)    
```
